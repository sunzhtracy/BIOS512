## Question 1
#### What is the holy trinity of Git? (Just name the parts.)
The holy trinity of Git consists of the branch, the working copy, and the staging area.

## Question 2
#### Explain how patches relate to the idea of committing a change to the history of your code?
Patches are the underlying concept behind how Git commits changes to the history of my code. A patch represents the difference between two versions of a file - what lines were added, removed, or modified. When I make a commit in Git, I am essentially saving a patch (a set of differences) that describes how to transform the previous state of my project into the new one. Over time, these patches form a continuous history of changes in my repository.

## Question 3
#### What is the difference between a branch and a fork?
A branch is a pointer to a specific series of commits within the same repository. It’s a temporary and flexible way to develop new features or experiments without affecting the main codebase. A fork, on the other hand, is a permanent copy of an entire repository that exists independently of the original.

## Question 4
#### How are diff and patch related? What’s the difference?
The diff command shows the differences between two versions of a file or set of files-it produces a “patch file” that records the changes needed to go from version A to version B. The patch command then takes that patch file and applies the recorded changes to transform one version into the other.

## Question 7
#### Walk me through how you would do interactive staging.
#### a) Firstly, what situation would interactive staging be useful in?
Interactive staging is useful when you’ve made several unrelated changes in the same file (or multiple files) but want to commit them separately with clear, focused commit messages. It lets you carefully select which specific parts (“hunks”) of your edits to include in each commit without having to undo or redo changes manually.

#### b) What git command would you use to start interactive staging?

#### c) On the text-based interactive menu, what option do you use?
On the text-based interactive menu that appears, you choose option p for “patch.” This allows you to review and select specific parts of changes (hunks) within a file.
#### d) After pressing that option, how do you select the file you want to stage?
After selecting p, Git will show a list of changed files and ask which one you want to review. You type the number corresponding to the file you want to stage, then press Enter.
#### e) What option do you type if you do not want to stage a hunk?
If you do not want to stage a hunk, you type n and press Enter.
#### f) Once you get to a hunk you want to stage, what do you do?
Once you get to a hunk you do want to stage, you type y and press Enter.
#### g) What if we have more hunks we want to stage?
